## Changepoint Analysis

```pyspark
#For Change Point Analysis
import ruptures as rpt
import numpy as np
```

```pyspark
# Australia Changepoint
plt.clf()
points=np.array(aus_mobility_data_final_pd["Result(Avg Non-Essential Activity Change)"].head(175))

#RUPTURES PACKAGE
#Changepoint detection with the Pelt search method
model="rbf"
algo = rpt.Pelt(model=model).fit(points)
result = algo.predict(pen=15)
rpt.display(points, result, figsize=(10, 6))
plt.title('Change Point Detection: Australia Non-Essential')
%matplot plt
```

```pyspark
# Find the entry numbers that change point identifies.
print(result[:-1])
```

```pyspark
# Use the entry numbers to find the date of the change point.
aus_mobility_data_final_pd.iloc[35].date
```

```pyspark
aus_mobility_data_final_pd.iloc[90].date
```
